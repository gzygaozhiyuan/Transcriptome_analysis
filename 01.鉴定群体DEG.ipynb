```python
import numpy as np
import pandas as pd
```

```python
# 读取ck表达矩阵
ck = pd.read_csv('./ck_ex_fin.txt', index_col=0, sep='\t', header=0)
ck.head()
```

```python
# 读取sa表达矩阵
sa = pd.read_csv('./sa_ex_fin.txt', index_col=0, sep='\t', header=0)
sa.head()
```

```python
# 读取22081基因列表
gene_ls = []
f1 = open('./22081gene.txt', 'r')
for line in f1:
    gene_ls.append(line.strip())
f1.close()
print(gene_ls[:5])
print(len(gene_ls))
```

```python
# 进行wilcox检验
from scipy.stats import wilcoxon
```

```python
# 创建一个空的dataframe，用于存储wilcox检验的结果，列名为Gene、Pvalue、log2FC
result = pd.DataFrame(columns=['Gene', 'Pvalue', 'CK(mean)','Salt(mean)','FC(mean)','CK(median)','Salt(median)','FC(median)'])
```

```python
# 用for循环进行wilcox检验
for i in gene_ls:
    ck1 = ck.loc[i]
    sa1 = sa.loc[i]
    pvalue = wilcoxon(ck1, sa1)[1]
    ck_mean = np.mean(ck1)
    sa_mean = np.mean(sa1)
    fc_mean = (sa_mean+1) / (ck_mean+1)
    ck_median = np.median(ck1)
    sa_median = np.median(sa1)
    fc_median = (sa_median+1) / (ck_median+1)
    result = result.append({'Gene':i, 'Pvalue':pvalue, 'CK(mean)':ck_mean, 'Salt(mean)':sa_mean, 'FC(mean)':fc_mean, 'CK(median)':ck_median, 'Salt(median)':sa_median, 'FC(median)':fc_median}, ignore_index=True)
result.head()
```

```python
# 对result的Pvalue进行校正
from statsmodels.stats.multitest import multipletests
```

```python
# 使用BH法进行校正
result['Pvalue_adj'] = multipletests(result['Pvalue'], method='fdr_bh')[1]
result.head()
```

```python
result_sig = result[result['Pvalue_adj']<0.05]
result_sig.head()
```

```python
# 保存result_sig
result_sig.to_csv('./1221result_sig.txt', sep='\t', header=True, index=True)
```

```python
# 保存结果
result.to_csv('./712_new_deg.txt', sep='\t', index=False)
result_sig.to_csv('./712_new_deg_sig.txt', sep='\t', index=False)
```

```python
# 给result添加一列Type，Pvalue_adj<0.05且log2FC>1的为up，Pvalue_adj<0.05且log2FC<-1的为down，其余为no
result['Type'] = 'no'
result.loc[(result['Pvalue_adj'] < 0.05) & (result['log2FC'] > 1), 'Type'] = 'up'
result.loc[(result['Pvalue_adj'] < 0.05) & (result['log2FC'] < -1), 'Type'] = 'down'
result.head()
```

```python
# 将result中Type为up的提取出来
up = result[result['Type'] == 'up']
up.shape
```

```python
# 将result中Type为down的提取出来
down = result[result['Type'] == 'down']
down.shape
```

```python
# 将result、up、down分别写入txt文件
result.to_csv('./DEG.txt', sep='\t', index=False, header=True)
up.to_csv('./up.txt', sep='\t', index=False, header=True)
down.to_csv('./down.txt', sep='\t', index=False, header=True)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 读取DEG.txt
deg = pd.read_csv('./DEG.txt', sep='\t', header=0)
deg.head()
```

```python
# 根据deg，绘制火山图
plt.figure(figsize=(8, 6))
plt.scatter(deg['log2FC'], -np.log10(deg['Pvalue_adj']), s=10, c='grey', alpha=0.5)
plt.scatter(deg[deg['Type'] == 'up']['log2FC'], -np.log10(deg[deg['Type'] == 'up']['Pvalue_adj']), s=10, c='red', alpha=0.5)
plt.scatter(deg[deg['Type'] == 'down']['log2FC'], -np.log10(deg[deg['Type'] == 'down']['Pvalue_adj']), s=10, c='blue', alpha=0.5)
# x=1和x=-1的竖线
plt.vlines(x=1, ymin=0, ymax=130, colors='black', linestyles='--', lw=1)
plt.vlines(x=-1, ymin=0, ymax=130, colors='black', linestyles='--', lw=1)
# 添加标题和坐标轴标签
plt.xlabel('log2FC')
plt.ylabel('-log10(Pvalue_adj)')
plt.savefig('./deg.pdf', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# 根据deg绘制柱状图
plt.figure(figsize=(8, 6))
sns.countplot(deg['Type'], order=['up', 'down'], palette=['red', 'blue'])
plt.text(0,823,ha='center', va='bottom', s='813',fontsize=15)
plt.text(1,175,ha='center', va='bottom', s='165',fontsize=15)
# 设置x轴和y轴字体大小
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
# 添加标题和坐标轴字体大小
plt.xlabel('Type', fontsize=15)
plt.ylabel('Count', fontsize=15)
plt.savefig('./deg_bar.pdf', dpi=300, bbox_inches='tight')
plt.show()
```
