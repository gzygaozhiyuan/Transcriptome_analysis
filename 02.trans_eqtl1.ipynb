```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Salt组

```python
salt_57w = pd.read_csv('./trans_sa_57w1_result1_all.txt', sep='\t', header=0)
salt_57w.head()
```

```python
salt_57w_unique = salt_57w.loc[salt_57w.groupby('rs')['p_wald'].idxmin()]
salt_57w_unique.head()
```

```python
# 按照chr和ps排序
salt_57w_unique = salt_57w_unique.sort_values(by=['chr', 'ps'])
salt_57w_unique.head()
```

```python
# 添加bin_label
salt_57w_unique['bin'] = None
bin_num = 1
prev_chr = None
prev_ps = None
for index, row in salt_57w_unique.iterrows():
    if prev_chr is None or row['chr'] != prev_chr or row['ps'] - prev_ps > 10000:
        bin_label = f'bin{bin_num}'
        bin_num += 1
        prev_chr = row['chr']
        prev_ps = row['ps']
    salt_57w_unique.at[index, 'bin'] = bin_label
salt_57w_unique.head()
```

```python
# 保存结果
salt_57w_unique.to_csv('./sa_each_snp_belong_to_which_bin.txt', sep='\t', index=False)
```

```python
# salt_57w_unique只保留最小的p值
salt_57w_unique_each_bin = salt_57w_unique.loc[salt_57w_unique.groupby('bin')['p_wald'].idxmin()]
salt_57w_unique_each_bin.head()
```

```python
# 按照chr和ps排序
salt_57w_unique_each_bin = salt_57w_unique_each_bin.sort_values(by=['chr', 'ps'])
salt_57w_unique_each_bin.head()
```

```python
# 保存结果
salt_57w_unique_each_bin.to_csv('./sa_each_bin_lead_snp.txt', sep='\t', index=False, header=True)
```

```python
# 读取trans
sa = pd.read_csv('./sa_1_trans_bin_leadsnp.txt',sep='\t', header=0)
sa.head()
```

```python
# 读取ld
sa_ld = pd.read_csv('./sa1_bin_re_ld.txt',sep='\t', header=0)
sa_ld.head()
```

```python
# 添加一列new_bin
sa['new_bin'] = None
new_bin_num = 1
pre_snp = None
for index, row in sa.iterrows():
    if pre_snp is None:
        pre_snp = row['rs']
        new_bin_label = f'new_bin{new_bin_num}'
        r2 = 1
        sa.at[index, 'new_bin'] = new_bin_label
    else:
        if row['rs'].split('_')[0][2:] == pre_snp.split('_')[0][2:]:
            r2_rows = sa_ld.loc[(sa_ld['snp1'] == pre_snp) & (sa_ld['snp2'] == row['rs']), 'r2']
            if len(r2_rows) > 0:
                r2 = r2_rows.values[0]
            else:
                r2 = 9
            if r2 > 0.1 and r2 != 9:
                sa.at[index, 'new_bin'] = new_bin_label
            elif r2 == 9:
                new_bin_num += 1
                new_bin_label = f'new_bin{new_bin_num}'
                sa.at[index, 'new_bin'] = new_bin_label
            else:
                new_bin_num += 1
                new_bin_label = f'new_bin{new_bin_num}'
                sa.at[index, 'new_bin'] = new_bin_label
        else:
            new_bin_num += 1
            new_bin_label = f'new_bin{new_bin_num}'
            sa.at[index, 'new_bin'] = new_bin_label
        pre_snp = row['rs']
sa.head()
```

```python
# 保存结果，空值用nan表示
sa.to_csv('./sa_each_snp_belong_to_which_new_bin0.01.txt', sep='\t', index=False, header=True, na_rep='NA')
```
