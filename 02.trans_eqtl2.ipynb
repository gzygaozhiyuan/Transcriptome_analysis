```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Salt组

```python
sa_bin = pd.read_csv('./sa_1_trans_snp2bin.txt', sep='\t', header=0)
sa_bin.head()
```

```python
sa_new_bin = pd.read_csv('./sa_each_snp_belong_to_which_new_bin0.01.txt', sep='\t', header=0)
sa_new_bin.head()
```

```python
# sa_new_bin每个new_bin保留最小p_wald的行
sa_new_bin_unique = sa_new_bin.loc[sa_new_bin.groupby('new_bin')['p_wald'].idxmin()]
sa_new_bin_unique.head()
```

```python
# 按照chr和ps排序
sa_new_bin_unique = sa_new_bin_unique.sort_values(by=['chr', 'ps'])
sa_new_bin_unique.head()
```

```python
# 给sa_bin添加新的列new_bin，值为sa_bin的bin列在sa_new_bin中的bin列对应new_bin列的值
sa_bin['new_bin'] = sa_bin['bin'].map(sa_new_bin.set_index('bin')['new_bin'])
sa_bin.head()
```

```python
# 保存sa_bin
sa_bin.to_csv('./sa_each_snp_belong_to_which_bin_fin.txt', sep='\t', index=False)
```

```python
# 保存sa_new_bin_unique
sa_new_bin_unique.to_csv('./sa_new_bin_lead_snp.txt', sep='\t', index=False)
```
